# Classifying Iris dataset using SVM

## Introduction 
Support Vector Machine (SVM) is a supervised machine learning algorithm that can be used for both classification or regression . However,  it is mostly used in classification problems. In this algorithm, we plot each data item as a point in n-dimensional space (where n is number of features you have) with the value of each feature being the value of a particular coordinate. Then, we perform classification by finding the hyper-plane that differentiate the two(multiple)classes very well .



## Loading the Data

This data sets consists of 3 different types of irisesâ€™ (Setosa, Versicolour, and Virginica) 
The rows being the samples and the columns(features) being: Sepal Length, Sepal Width, Petal Length and Petal Width.
First we have to import 'load_iris' to load the dataset .


```python
from sklearn.datasets import load_iris
# loading the dataset
iris = load_iris()
features = iris.data
target = iris.target

print("Data loaded successfuly and the number of samples is: ", iris.target.size)
```

## Data Exploration 
### 2D Plotting
Here we plot each two of our features against the classes to discover how they relate to each class.Diffrent colors denote (of course :D) different classes.


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets


x = iris.data 
y = iris.target

fig = plt.figure(figsize=(20,40))

subplots = fig.add_subplot(6,2,1)
subplots.set_xlabel('Sepal Length')
subplots.set_ylabel('Sepal Width')
subplots.scatter(x[:,0],x[:,1], c=y)

subplots = fig.add_subplot(6,2,2)
subplots.set_xlabel('Sepal Length')
subplots.set_ylabel('Petal Length')
subplots.scatter(x[:,0],x[:,2], c=y)

subplots = fig.add_subplot(6,2,3)
subplots.set_xlabel('Sepal Length')
subplots.set_ylabel('Petal Width')
subplots.scatter(x[:,0],x[:,3], c=y)

subplots = fig.add_subplot(6,2,4)
subplots.set_xlabel('Sepal Width')
subplots.set_ylabel('Petal Length')
subplots.scatter(x[:,1],x[:,2], c=y)

subplots = fig.add_subplot(6,2,5)
subplots.set_xlabel('Sepal Width')
subplots.set_ylabel('Petal Width')
subplots.scatter(x[:,1],x[:,3], c=y)

subplots = fig.add_subplot(6,2,6)
subplots.set_xlabel('Petal Length')
subplots.set_ylabel('Petal Width')
subplots.scatter(x[:,2],x[:,3], c=y)

```

### 3D Plots
Here we plot each three of our four features against the classes to discover how they relate to each class.Again Diffrent colors denote (of course :D) different classes.

```python
fig = plt.figure(figsize=(20,60))

subplot = fig.add_subplot(411, projection='3d')
subplot.scatter(x[:,1],x[:,2], x[:,3], c=y)

subplot = fig.add_subplot(412, projection='3d')
subplot.scatter(x[:,0],x[:,1], x[:,2], c=y)

subplot = fig.add_subplot(413, projection='3d')
subplot.scatter(x[:,0],x[:,1], x[:,3], c=y)

subplot = fig.add_subplot(414, projection='3d')
subplot.scatter(x[:,0],x[:,2],x[:,3], c=y)

```

## Data Splitting 
Splitting the dataset for training and testing. 80% percent of the data goes for training, every datapoint in X_train with its correspending output in y_train are used to train our model. The remaining 20% of data is used afterwards to test the model on unseen data.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=15)
print("Data Splitted Successfully !")
```
